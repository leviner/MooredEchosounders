```python
import pandas as pd
import matplotlib.pyplot as plt
def readTracks(file):
    dfTracks = pd.read_csv(file)
    dfTracks['Date_S'] = dfTracks['Date_S'].astype(str)
    dfTracks['Datetime'] = pd.to_datetime(dfTracks['Date_S'] + ' ' + dfTracks['Time_S'])
    dfTracks = dfTracks.drop(['Date_S','Time_S'], axis=1)
    dfTracks = dfTracks.set_index('Datetime')
    return dfTracks

df1 = readTracks('../data/analysisFiles/TrackRegions6.csv')
df2 = readTracks('../data/analysisFiles/TrackRegions6_linear2.csv')
```

```python
plt.hist(df1.directHead,bins=90,alpha=0.5)
plt.hist(df2.directHead,bins=90,alpha=0.5);
```

```python
plt.hist(df1.directHead,bins=90,alpha=0.5)
plt.hist(df2.linearHead,bins=90,alpha=0.5);
```

```python
plt.plot(df2.linearHead,df2.directHead,'.')
plt.gca().set_aspect('equal', adjustable='box')
plt.axis([0,360,0,360])
plt.plot([0,360],[0,360])
plt.xlabel('Linear Heading')
plt.ylabel('Spline Heading')
```

```python
plt.plot(df2.linearSpeed,df2.directSpeed,'.')
plt.gca().set_aspect('equal', adjustable='box')
plt.plot([0,5],[0,5])
plt.axis([0,5,0,5])
plt.xlabel('Linear Speed')
plt.ylabel('Spline Speed')
```
